```python
import requests as req 
```

```python
#URL for Miso's Day Ahead Wind Forecast API

res = req.get('https://api.misoenergy.org/MISORTWDDataBroker/DataBrokerServices.asmx?messageType=getWindForecast&returnType=json')
```

```python
res.json()['Forecast']
```

```python
import pandas as pd
```

```python
days = res.json()['Forecast']
```

```python
days[0]
```

```python
df = pd.io.json.json_normalize(days[0])
```

```python
df.head()
```

```python
for day in days[1:]:
    df = df.append(pd.io.json.json_normalize(day))
```

```python
df.shape
```

```python
df
```

```python
df.info()
```

```python
df['DateTimeEST'] = pd.to_datetime(df['DateTimeEST'], utc=True)
```

```python
df.info()
```

```python
df.drop(['HourEndingEST'], inplace=True, axis=1)
```

```python
df.info()
```

```python
df
```
